## 実験F コイン投げによる勝ち負けのプロセス
### F-1 コイン投げの実験
①コイン投げの2つの事象を起こし, 結果を1(オモテ)と0(ウラ)として表示するプログラムを作成する.

```r
x = runif(1)
if (x>=0.5){
    x =1
}else{
    x=0
}
print(x)
```

②n回繰り返したときの各事象の発生する確率を求めなさい.

```r
n = 100
count = 0
for(i in 1:100){
        x = runif(n)[i]
        if (x>=0.5){
            x =1
            count = count + 1
        }else{
            x=0
        }
        }
print("オモテ")
print(count)
print("ウラ")
print(n - count)
```

### F-2 確率過程の設計
①オモテが出た場合は+1を, ウラが出た場合は-1として, n回繰り返すシミュレーション実験を設計する.  
その際に, 各回での得点を記録(表示)するようにしなさい.

```r
n = 100
point    = 0
point[1] = 0
for(i in 1:n){
        x = runif(n)[i]
        if (x>=0.5){
            x =1
            point[i+1] = point[i] + 1 
        }else{
            x=-1
            point[i+1] = point[i] - 1
        }
#         print(point[i])
        }
plot(point, type="l")
abline(0, 0, col="red")
```

## 実験H 釣銭準備の最適化問題
「鯛焼き」一個70円で, 500個販売する. 10円玉を何枚準備しておけばいいか？
①問題を把握して, 状況を抽象化, モデル化する.    
→お客が出してくる金のパターンは    
(イ) 100 + 10 × 2 円 → +2枚  
(ロ) 100 円 → -3枚  
(ハ) 50 + 50 円 → -3枚  
(ニ) 10 × 7 円 → +7枚  
(ホ) 50 + 10 × 2 円 → +2枚  
②モデルをもとに, シミュレーション実験の設計を行う.  
→**(イ)～(ホ)は等確率で発生するケースとして, ランダムウォークを適用する.**  
例えば, runif()の出力値が0≦x＜0.2なら+2, 0.2≦x＜0.4なら-3, といったようにすればよい.


```r
n = 100
x = runif(n)
coin = 0
coin[1] = 0
for ( i in 1:n){
    if(x[i] < 0.6){
        if(x[i] < 0.4){
            coin[i+1] = coin[i] + 2
        }else{
            coin[i+1] = coin[i] + 2
        }
    }else{
        coin[i+1] = coin[i] -3
    }
    print(coin[i])
}
```
